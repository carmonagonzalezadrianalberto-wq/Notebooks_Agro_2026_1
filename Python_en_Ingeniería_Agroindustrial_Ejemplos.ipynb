### Ejemplos Sencillos de Código

Aquí tienes tres aplicaciones prácticas donde Python simplifica la vida en el campo y la planta.

#### 1. Cálculo de Rendimiento de Extracción

Este pequeño script ayuda a determinar qué porcentaje de materia prima se convirtió realmente en producto terminado (por ejemplo, aceite de palma o jugo de fruta).

```
materia_prima = 500  # kg ingresados
producto_final = 120 # kg obtenidos

rendimiento = (producto_final / materia_prima) * 100

print(f"El rendimiento de la extracción es del {rendimiento}%")
```

#### 2. Monitoreo de Temperatura de Almacenamiento

Ideal para asegurar que la cadena de frío no se rompa.

```
temperatura_silo = 22.5  # Dato que vendría de un sensor

if temperatura_silo > 18.0:
    print("ALERTA: Temperatura fuera de rango. Activando refrigeración...")
else:
    print("Temperatura óptima. Calidad preservada.")
```

#### 3. Cálculo de Humedad en Granos

Un cálculo básico para decidir si un lote de grano está listo para su procesamiento o venta.

```
peso_humedo = 10.5
peso_seco = 8.2

humedad_base_humeda = ((peso_humedo - peso_seco) / peso_humedo) * 100

print(f"La humedad del grano es: {humedad_base_humeda:.2f}%")
```
